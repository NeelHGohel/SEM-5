### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
data_minmax = df.copy()
age_mn = data_minmax['Age']
age_min = age_mn.min()
age_max = age_mn.max()
data_minmax['MinMaxAge'] = (age_mn - age_min)/(age_max - age_min)
data_minmax[['Age','MinMaxAge']]
```
